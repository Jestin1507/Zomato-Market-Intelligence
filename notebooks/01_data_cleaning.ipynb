- Total Records: 51,717
- Total Features: 17
- Dish Liked contains 54.29% missing values.
- Ratings contain 15.03% missing values.
- Remaining features have less than 3% missing values.
- Dataset is suitable for exploratory and predictive analysis after preprocessing.

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/raw/zomato.csv")

df.head()
```

```python
print("Shape:", df.shape)

df.info()
```

```python
missing_values = df.isnull().sum()

missing_values.sort_values(ascending=False)
```

```python
missing_percentage = (
    df.isnull().sum() / len(df)
) * 100

missing_percentage.sort_values(ascending=False)
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
df = df.drop_duplicates()

print(df.shape)
```

```python
df["rate"].unique()[:20]
```

```python
df["rate"] = df["rate"].replace(
    ["NEW", "-"],
    np.nan
)

df["rate"] = (
    df["rate"]
    .astype(str)
    .str.replace("/5", "", regex=False)
)

df["rate"] = pd.to_numeric(
    df["rate"],
    errors="coerce"
)
```

```python
df["rate"].head()
```

```python
df["approx_cost(for two people)"] = (
    df["approx_cost(for two people)"]
    .astype(str)
    .str.replace(",", "")
)

df["approx_cost(for two people)"] = pd.to_numeric(
    df["approx_cost(for two people)"],
    errors="coerce"
)
```

```python
df.info()
```

```python
df.to_csv(
    "../data/processed/cleaned_zomato.csv",
    index=False
)

print("Cleaned dataset saved successfully!")
```

```python
analysis_df = df.copy()

analysis_df = analysis_df.dropna(
    subset=[
        "location",
        "cuisines",
        "approx_cost(for two people)"
    ]
)

analysis_df["rest_type"] = (
    analysis_df["rest_type"]
    .fillna("Unknown")
)

print(analysis_df.shape)
```

```python
analysis_df.to_csv(
    "../data/processed/analysis_zomato.csv",
    index=False
)
```

```python
master_df = analysis_df.copy()

master_df = master_df.drop_duplicates(
    subset=["name", "address"]
)

print(master_df.shape)
```

```python
master_df.to_csv(
    "../data/processed/master_zomato.csv",
    index=False
)
```
