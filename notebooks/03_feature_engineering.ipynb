```python
import pandas as pd
import numpy as np

df = pd.read_csv(
    "../data/processed/master_zomato.csv"
)
```

```python
df["votes_norm"] = (
    df["votes"] - df["votes"].min()
) / (
    df["votes"].max() - df["votes"].min()
)
```

```python
df["online_order_num"] = (
    df["online_order"] == "Yes"
).astype(int)

df["book_table_num"] = (
    df["book_table"] == "Yes"
).astype(int)
```

```python
df["success_score"] = (
    0.5 * df["rate"].fillna(df["rate"].median())
    +
    0.3 * df["votes_norm"]
    +
    0.1 * df["online_order_num"]
    +
    0.1 * df["book_table_num"]
)
```

```python
df[
    ["name",
     "location",
     "rate",
     "votes",
     "success_score"]
].sort_values(
    "success_score",
    ascending=False
).head(20)
```
