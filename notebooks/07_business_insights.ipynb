## Business Insight 1: Restaurant Density

BTM, HSR and Koramangala have the highest concentration of restaurants in Bengaluru.

This indicates:

- Strong customer demand
- High restaurant competition
- Attractive markets for food businesses

New restaurants entering these locations should focus on differentiation due to market saturation.

```python
import plotly.io as pio

pio.renderers.default = "browser"
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv(
    "../data/processed/final_zomato.csv"
)

print(df.shape)

df.head()
```

```python
top_locations = (
    df["location"]
    .value_counts()
    .head(10)
    .reset_index()
)

top_locations.columns = [
    "Location",
    "Restaurant_Count"
]

top_locations
```

```python
fig = px.bar(
    top_locations,
    x="Restaurant_Count",
    y="Location",
    orientation="h",
    text="Restaurant_Count",
    color="Restaurant_Count",
    color_continuous_scale="Reds"
)

fig.update_layout(
    title="Top 10 Restaurant Locations",
    yaxis=dict(
        categoryorder="total ascending"
    )
)

fig.show()
```

## Business Insight 2: Cuisine Demand

North Indian cuisine dominates Bengaluru's restaurant market, followed by Chinese and South Indian cuisine.

Key Findings:

- North Indian food has the highest market demand.
- Chinese cuisine has widespread adoption across locations.
- South Indian cuisine remains a core category despite increasing global food trends.
- Fast Food and Biryani continue to attract significant customer interest.

Business Impact:

New restaurant owners entering Bengaluru should prioritize high-demand cuisines or differentiate strongly when entering saturated cuisine categories.

```python
cuisine_counts = (
    df["cuisines"]
    .dropna()
    .str.split(",")
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
    .reset_index()
)

cuisine_counts.columns = [
    "Cuisine",
    "Count"
]

cuisine_counts
```

```python
fig = px.bar(
    cuisine_counts,
    x="Count",
    y="Cuisine",
    orientation="h",
    text="Count",
    color="Count",
    color_continuous_scale="Reds"
)

fig.update_layout(
    title="Top 10 Most Popular Cuisines",
    yaxis=dict(
        categoryorder="total ascending"
    )
)

fig.show()
```

## Business Insight 3: Highest Rated Locations

Lavelle Road, Koramangala 3rd Block and St. Marks Road have the highest average restaurant ratings in Bengaluru.

### Key Findings

- Premium locations consistently receive higher customer ratings.
- Customers in these areas are willing to pay more for quality dining experiences.
- These locations demonstrate strong restaurant performance and customer satisfaction.

### Business Recommendation

Restaurant investors seeking premium markets should prioritize:

- Lavelle Road
- Koramangala
- Church Street
- St. Marks Road

These areas offer opportunities for high-value dining experiences and strong customer engagement.

```python
location_rating = (
    df.groupby("location")["rate"]
    .mean()
    .sort_values(
        ascending=False
    )
    .head(10)
    .reset_index()
)

location_rating
```

```python
fig = px.bar(
    location_rating,
    x="rate",
    y="location",
    orientation="h",
    text="rate",
    color="rate",
    color_continuous_scale="Greens"
)

fig.update_traces(
    texttemplate="%{text:.2f}",
    textposition="outside"
)

fig.update_layout(
    title="Top Rated Restaurant Locations",
    yaxis=dict(
        categoryorder="total ascending"
    )
)

fig.show()
```

## Business Insight 4: Online Ordering Impact

Restaurants offering online ordering achieve higher average ratings compared to restaurants that do not provide online ordering.

### Key Findings

- Online Ordering Restaurants: ~3.72
- Non Online Ordering Restaurants: ~3.66

### Business Recommendation

Providing online ordering improves customer convenience and accessibility, which contributes to higher customer satisfaction and ratings.

Restaurants seeking growth should prioritize digital ordering platforms and delivery integration.

```python
online_rating = (
    df.groupby("online_order")["rate"]
    .mean()
    .reset_index()
)

online_rating
```

```python
fig = px.bar(
    online_rating,
    x="online_order",
    y="rate",
    text="rate",
    color="online_order",
    title="Impact of Online Ordering on Ratings"
)

fig.update_traces(
    texttemplate="%{text:.2f}",
    textposition="outside"
)

fig.show()
```
